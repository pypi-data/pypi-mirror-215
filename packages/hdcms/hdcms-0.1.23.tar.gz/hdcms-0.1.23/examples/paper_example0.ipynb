```python
import hdcms as hdc
import matplotlib.pyplot as plt
from PIL import Image
```

```python
sum1 = hdc.regex2stats2d("1-\d.txt", dir="./data/DART-MS/30V_Centroid", xtol=0.1)
sum2 = hdc.regex2stats2d("2-\d.txt", dir="./data/DART-MS/30V_Centroid", xtol=0.1)
print(sum1)
```

```python
image_config = hdc.visualize.ImageConfig(size=(3600, 1200), std_scale=1, desingularization=1e-2, axis_thickness=3, xlabel="x values", ylabel="y values")
# axis = (min_x, min_y, max_x, max_y)
axis = (0, -0.1, 400, 1.2)
```

```python
_ = hdc.write_image(sum1, config=image_config, axis_limits=axis)
```

```python
hdc.compare(sum1, sum2)
```
