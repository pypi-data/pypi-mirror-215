Because they are a 9000 element array of pairs, it's hard too see what's going on. We can tell that most of the elements are 0. We can visualize the summary statistic better using the following.

```python
image_config = hdc.visualize.ImageConfig(size=(3600, 1200), std_scale=1, desingularization=1e-2, axis_thickness=3, xlabel="m/z values", ylabel="intensity")
# axis = (min_x, min_y, max_x, max_y)
axis = (170, -0.04, 180, 0.25)
```

```python
img = hdc.write_image(sum1, config=image_config, axis_limits=axis)
```
