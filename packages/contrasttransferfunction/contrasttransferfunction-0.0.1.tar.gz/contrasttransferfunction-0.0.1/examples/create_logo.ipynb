```python
import matplotlib.pyplot as plt

golden_ratio = (1 + 5 ** 0.5) / 2

# Set figure size to golden ratio
fig_width = 8
fig_height = fig_width / golden_ratio
fig, ax = plt.subplots(figsize=(fig_width, fig_height))


# Add text
text = ax.text(0.5, 0.5, 'contrasttransferfunction', ha='center', va='center', fontsize=36, fontname='Roboto')

# Set axis limits and remove ticks
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_xticks([])
ax.set_yticks([])

# Save figure
plt.savefig('logo.svg')
```
